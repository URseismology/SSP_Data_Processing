```python
# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# Author: Tolulope Olugboji
# Date:  October 21, 2013
# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
#
# Objective:   Prepare Manifest List [For RFAnalysis] for All SAC Data from Kawakatsu
#              Plot Station-Event Information by crawling all directories
#              Do Azimuth Rose Plot .. 
#              .. Other statistics analysois
#
#
# Code uses T1 header to create QC Manifest ...
#
#
from obspy import UTCDateTime
from obspy import read
import obspy.signal
import time, sys, os
from IPython.display import display, clear_output
from obspy.taup.taup import getTravelTimes

def readHeaders(trace):
    "Read relevant header information ..."
    return {'gcarc': trace.stats.sac['gcarc'], 'evdp':trace.stats.sac['evdp'], 
            'starttime':trace.stats['starttime'], 'o': trace.stats['sac']['o'], 'delta': trace.stats['delta'],
            'baz':trace.stats.sac['baz'], 't1':trace.stats.sac['t1']}

########################### END OF FUNCTION DEFINITIONS #########################################################
#!!!!!!!!!!!!!!!!!!!!!!!!!!! BEGIN WORKFLOW HERE ...... !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

inDir = "/Users/tmo22/Documents/JapaneseProject/ProcData/"
projectDirs = ['sspData/', 'boreHole/']
yearDirs = ['2005', '2006', '2007']
boreHoleStations = ['WP1', 'WP2']

manifestDir = "/Users/tmo22/Documents/JapaneseProject/Manifest/"

for projectDir in projectDirs:
    
    if ( projectDir == 'sspData/'):
        
        ################################################################   1. Manifest by Year
        for yearDir in yearDirs:
        
            projectYearDir = inDir+projectDir+yearDir+'/' 
            # check stations and parse files in stations...
            stations = !ls $projectYearDir
           
            ###########################################################   1. Manifest by Station
            for station in stations:
                fileDir = projectYearDir+station+'/'
                #print fileDir
                
                fileNames = !ls $fileDir*Z
                
                # Open fileManifest and dump file names in fileManifest ...
                manifestName = yearDir+station+'.txt'
                manifestAllURL = manifestDir+'All/'+manifestName 
                manifestPassURL = manifestDir+'Pass/'+manifestName 
                manifestFailURL = manifestDir+'Fail/'+manifestName 
                
                # Do directory check and delete/create files as needed
                if ( os.path.exists(manifestAllURL)  ):
                    print "All file already exists, so delete and rewrite ..."
                    os.remove(manifestAllURL)
                    #file4All = open(manifestAllURL, 'wb')
                elif ( os.path.exists(manifestPassURL)  ):
                    print "Pass file already exists, so delete and rewrite ..."
                    os.remove(manifestPassURL)
                    #file4Pass = open(manifestPassURL, 'wb')
                elif ( os.path.exists(manifestFailURL)  ):
                    print "Fail file already exists, so delete and rewrite ..."
                    os.remove(manifestFailURL)
                    #file4Fail = open(manifestFailURL, 'wb')   
                else:
                    print "no manifest with this name, creating now ..."
                file4All = open(manifestAllURL, 'wb')
                file4Pass = open(manifestPassURL, 'wb')
                file4Fail = open(manifestFailURL, 'wb')
                
                cntQCPass = 0
                cntQCFail = 0
                ###########################################################   1. Run through Records ..
                #for fileName in fileNames[0:1]:
                for fileName in fileNames:
                    
                    vertStrm = read(fileName);
                    sacDics = readHeaders( vertStrm[0] )
                    
                    # update Manifest and other 
                    lenName = len(fileName) - 1
                    file4All.write(fileName[0:lenName]+'\n')
                    
                    if (sacDics['t1'] > 0 ):
                        cntQCPass += 1
                        #print "Total Pass: ", cntQCPass
                        file4Pass.write(fileName[0:lenName]+'\n')
                    else:
                        cntQCFail += 1
                        #print "Total Not Pass: ", cntQCFail
                        file4Fail.write(fileName[0:lenName]+'\n')
                
                    print fileDir, "Pass: ", cntQCPass, "Fail: ", cntQCFail
                    clear_output()
                
                # Close file manifest
                file4Fail.close()
                file4Pass.close()
                file4All.close()
                            
            
            
    
    else:
        
        for boreHoleStation in boreHoleStations:
            fileDir = inDir+projectDir+boreHoleStation+'/'
            #print fileDir
            
            fileNames = !ls $fileDir*Z
            
                           
            # Open fileManifest and dump file names in fileManifest ...
            manifestName = boreHoleStation+'.txt'
            manifestAllURL = manifestDir+'All/'+manifestName 
            manifestPassURL = manifestDir+'Pass/'+manifestName 
            manifestFailURL = manifestDir+'Fail/'+manifestName 
                
            # Do directory check and delete/create files as needed
            if ( os.path.exists(manifestAllURL)  ):
                print "file already exists, so delete and rewrite ..."
                os.remove(manifestAllURL)
                #file4All = open(manifestAllURL, 'wb')
            elif ( os.path.exists(manifestPassURL)  ):
                print "file already exists, so delete and rewrite ..."
                os.remove(manifestPassURL)
                #file4Pass = open(manifestPassURL, 'wb')
            elif ( os.path.exists(manifestFailURL)  ):
                print "file already exists, so delete and rewrite ..."
                os.remove(manifestFailURL)
                #file4Fail = open(manifestFailURL, 'wb')   
            else:
                print "no manifest with this name, creating now ..."
            file4All = open(manifestAllURL, 'wb')
            file4Pass = open(manifestPassURL, 'wb')
            file4Fail = open(manifestFailURL, 'wb')
                
            cntQCPass = 0
            cntQCFail = 0
            ###########################################################   1. Run through Records ..
            #for fileName in fileNames[0:1]:
            for fileName in fileNames:
                    
                vertStrm = read(fileName);
                sacDics = readHeaders( vertStrm[0] )
                    
                # update Manifest and other 
                lenName = len(fileName) - 1
                file4All.write(fileName[0:lenName]+'\n')
                    
                if (sacDics['t1'] > 0 ):
                    cntQCPass += 1
                    #print "Total Pass: ", cntQCPass
                    file4Pass.write(fileName[0:lenName]+'\n')
                else:
                    cntQCFail += 1
                    #print "Total Not Pass: ", cntQCFail
                    file4Fail.write(fileName[0:lenName]+'\n')
                
                print fileDir, "Pass: ", cntQCPass, "Fail: ", cntQCFail
                clear_output()
                
            # Close file manifest
            file4Fail.close()
            file4Pass.close()
            file4All.close()
                
```

```python
# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# Author: Tolulope Olugboji
# Date:  October 30, 2013
# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
#
# Objective:   Parse Manifest List [For RFAnalysis] for All SAC Data from Kawakatsu
#              Plot Station-Event Information by parsing manifest information ...
#              Do Azimuth Rose Plot .. 
#              .. Other statistics analysois
#
#
# Code uses T1 header to create QC Manifest ...
#
#

from obspy import UTCDateTime
from obspy import read
import obspy.signal
import time, sys, os
from IPython.display import display, clear_output
from obspy.taup.taup import getTravelTimes

import numpy as np
import matplotlib.pyplot as plt
import datetime
from mpl_toolkits.basemap import Basemap, shiftgrid
from netCDF4 import Dataset
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.backends.backend_pdf import PdfPages

def showPolarEventPlot(evntBas, evntDist, area, evMag, pltTitle, fName):
    #areaMag.append(3**eqHeaders['mag'])
    ax = plt.subplot(1,1,plotIndex, projection='polar')
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)
    c = plt.scatter((np.pi/180)*np.array(evBaz), np.array(evDist),  
                            s=np.array(areaMag), c= np.array(evMag), cmap=plt.cm.hot)
    c.set_alpha(0.75)
    cbar = plt.colorbar(c)
    cbar.ax.set_ylabel('Earthquake Magnitude')
    
    #plotIndex += 1
    #plt.show()
    #pp = PdfPages(imagesDir+'OBSstats.png')
    plt.title(pltTitle)
    plt.savefig(imagesDir+fName+'.png', format='png')
    plt.show()
    # Close Multipage file after done.
    #pp.close
    

def showStationEventMap(stCords, evCords, areaMag, evMag, pltTitle, fName):
    # Import utilities ... basemap, gridplot, netCDF4, etc.
    from mpl_toolkits.basemap import Basemap, shiftgrid, cm
    import numpy as np
    import matplotlib.pyplot as plt
    from netCDF4 import Dataset
    
    # read in etopo5 topography/bathymetry.
    url = 'http://ferret.pmel.noaa.gov/thredds/dodsC/data/PMEL/etopo5.nc'
    etopodata = Dataset(url)
    
    topoin = etopodata.variables['ROSE'][:]
    lons = etopodata.variables['ETOPO05_X'][:]
    lats = etopodata.variables['ETOPO05_Y'][:]
    # shift data so lons go from -180 to 180 instead of 20 to 380.
    topoin,lons = shiftgrid(180.,topoin,lons,start=False)

    plt.figure(figsize=(11,8))
    # lon_0, lat_0 are the center point of the projection.
    # resolution = 'l' means use low resolution coastlines.
    m = Basemap(projection='ortho',lon_0=-180,lat_0=0,resolution='l')
    
    # transform to nx x ny regularly spaced 5km native projection grid
    nx = int((m.xmax-m.xmin)/5000.)+1; ny = int((m.ymax-m.ymin)/5000.)+1
    topodat = m.transform_scalar(topoin,lons,lats,nx,ny, masked=True)
    # plot image over map with imshow.

    m.drawcoastlines()
    im = m.imshow(topodat,cm.GMT_haxby)
    #cs = m.contourf(lon,lat,topoin,cmap=m.GMT_haxby)
    #m.fillcontinents(color='coral',lake_color='aqua')
    
    #Scatter station coords ..
    evLonLat = np.array(evCords)
    stLonLat = np.array(stCords)
    
    x, y = m(evLonLat[:,0], evLonLat[:,1])
    xsta, ysta = m(stLonLat[:,0], stLonLat[:,1])

    
    c = m.scatter(x, y, s=10, c= np.array(evMag), cmap=plt.cm.hot)
    c.set_alpha(0.75)
    cbar = plt.colorbar(c)
    m.scatter(xsta, ysta, marker='^', s=50, c= 'red')
    
    # draw parallels and meridians.
    # m.drawparallels(np.arange(-90.,120.,30.))
    #m.drawmeridians(np.arange(0.,420.,60.))
    #m.drawmapboundary(fill_color='aqua')
    plt.title(pltTitle)
    plt.savefig(imagesDir+fName+'.png', format='png', dpi=600)
    plt.show()
    


def showHistogramScatterPlot(evntBaz, evntDist, area, evMag,  pltTitle, fName):
    # the random data
    x = np.array(evntBaz)
    y = np.array(evntDist)
    
    
    fig, axScatter = plt.subplots(figsize=(8.5,8.5))
    
    # the scatter plot:
    mapHolder = axScatter.scatter(x, y, s = np.array(area), c= np.array(evMag), cmap=plt.cm.hot)
    #axScatter.set_aspect(1.)
    plt.xlabel("Back Azimuth")
    plt.ylabel("Great Circle Distance")
    
    
    # create new axes on the right and on the top of the current axes
    # The first argument of the new_vertical(new_horizontal) method is
    # the height (width) of the axes to be created in inches.
    divider = make_axes_locatable(axScatter)
    axHistx = divider.append_axes("top", 1.2, pad=0.1, sharex=axScatter)
    plt.title(pltTitle)
    axHisty = divider.append_axes("right", 1.2, pad=0.1, sharey=axScatter)
    
    # make some labels invisible
    plt.setp(axHistx.get_xticklabels() + axHisty.get_yticklabels(),
             visible=False)
    
    # now determine nice limits by hand:
    binwidth = 10
    xymax = np.max( [np.max(np.fabs(x)), np.max(np.fabs(y))] )
    lim = ( int(xymax/binwidth) + 1) * binwidth
    
    bins = np.arange(-lim, lim + binwidth, binwidth)
    axHistx.hist(x, bins=bins)
    axHisty.hist(y, bins=bins, orientation='horizontal')
    
    # the xaxis of axHistx and yaxis of axHisty are shared with axScatter,
    # thus there is no need to manually adjust the xlim and ylim of these
    # axis.
    
    #axHistx.axis["bottom"].major_ticklabels.set_visible(False)
    for tl in axHistx.get_xticklabels():
        tl.set_visible(False)
    #axHistx.set_yticks([0, 20, 50])
    
    #axHisty.axis["left"].major_ticklabels.set_visible(False)
    for tl in axHisty.get_yticklabels():
        tl.set_visible(False)
    #axHisty.set_xticks([0, 20, 50])

    plt.savefig(imagesDir+fName+'.png', format='png')

    plt.draw()
    plt.show()

def readHeaders(trace):
    "Read relevant header information ..."
    return {'gcarc': trace.stats.sac['gcarc'], 'evdp':trace.stats.sac['evdp'], 
            'stla': trace.stats['sac']['stla'], 'stlo': trace.stats['sac']['stlo'], 
            'evla': trace.stats['sac']['evla'], 'evlo': trace.stats['sac']['evlo'],
            'baz':trace.stats.sac['baz'], 'mag':trace.stats.sac['mag']}

################   WORKFLOW Here ... Get Lat Lon Data ...
manifestDir = "/Users/tmo22/Documents/JapaneseProject/Manifest/"
imagesDir = "/Users/tmo22/Documents/JapaneseProject/Images/"
DataQuals = ['All/', 'Pass/', 'Fail/']



for DataQual in DataQuals[1:2]:
    loadManifest = manifestDir+DataQual
    manifestFiles = !ls $loadManifest
    
    # Save statistics for each manifest in a multipage pdf page . Close when done.
    
    stCords2005 = []; evCords2005 = []; evMag2005 = []; areaMag2005 = []
    stCords2006 = []; evCords2006 = []; evMag2006 = []; areaMag2006 = []
    stCords2007 = []; evCords2007 = []; evMag2007 = []; areaMag2007 = []
    
    stCordsBH = []; evCordsBH = []; evMagBH = []; areaMagBH = []
    
    for manifestFile in manifestFiles:
        if (manifestFile == 'WP1.txt' or manifestFile == 'WP2.txt') :
            
            
            openManifest = open(loadManifest+manifestFile, 'r'); 
            manifestTable = openManifest.readlines()
                        
            stCords = []
            evCords = []
            evMag = []; areaMag = []
            evBaz = []
            evDp = []
            evDist = []
            plotIndex = 1
            
            print "borehole", manifestFile, "Total Equakes: ", len(manifestTable)
            for fileName in manifestTable:
                #stream = read(fileName+'Z')
                fileHead = fileName.split("\n")
                stream = read(fileHead[0]+'Z')
                eqHeaders = readHeaders(stream[0])
                
                stCords.append( [eqHeaders['stlo'], eqHeaders['stla'] ] )
                evCords.append( [eqHeaders['evlo'], eqHeaders['evla'] ] )
                evCordsBH.append( [eqHeaders['evlo'], eqHeaders['evla'] ] )
                
                evMag.append( eqHeaders['mag'] )
                evMagBH.append( eqHeaders['mag'] )
                
                evBaz.append( eqHeaders['baz'] )
                evDp.append( eqHeaders['evdp'] )
                evDist.append( eqHeaders['gcarc'] )
            
                # Done harvesting the event headers, now do statistics / display ...
                areaMag.append( np.pi * (eqHeaders['mag']**2) )# 0 to Magnitude point radiuses
                areaMagBH.append( np.pi * (eqHeaders['mag']**2) )# 0 to Magnitude point radiuses
                
            stCordsBH.append( [eqHeaders['stlo'], eqHeaders['stla'] ] )
        
            
            # Do polar Plot .........
            manifestHead = manifestFile.split("."); manifestHead = manifestHead[0]
            plotFName = 'StatsBorehole/' + DataQual +  "Polar/" +  manifestHead
            plotTitle = "BoreHole: " + manifestHead + " | Total Equakes: " + str( len(manifestTable) ) + " | QC Tag: " + DataQual

            #showPolarEventPlot(evBaz, evDist, areaMag, evMag, plotTitle, plotFName)
            
            # Do scatter histogram plot
            plotFName = 'StatsBorehole/' + DataQual +  "Histograms/" +  manifestHead
            plotTitle = "BoreHole: " + manifestHead + " | Total Equakes: " + str( len(manifestTable) ) + " | QC Tag: " + DataQual

            #showHistogramScatterPlot(evBaz, evDist, areaMag, evMag, plotTitle, plotFName)
            
            

        else:
            
            openManifest = open(loadManifest+manifestFile, 'r'); 
            manifestTable = openManifest.readlines()
            
            stCords = []
            evCords = []
            evMag = []; areaMag = []
            evBaz = []
            evDp = []
            evDist = []
            plotIndex = 1
            
            
            
            print "OBS", manifestFile, "Total Equakes: ", len(manifestTable)
            for fileName in manifestTable:
                #break
                #stream = read(fileName+'Z')
                fileHead = fileName.split("\n")
                stream = read(fileHead[0]+'Z')
                eqHeaders = readHeaders(stream[0])
                
                stCords.append( [eqHeaders['stlo'], eqHeaders['stla'] ] )
                evCords.append( [eqHeaders['evlo'], eqHeaders['evla'] ] )
                
                evMag.append( 5.0 )
                evBaz.append( eqHeaders['baz'] )
                evDp.append( eqHeaders['evdp'] )
                evDist.append( eqHeaders['gcarc'] )
            
                # Done harvesting the event headers, now do statistics / display ...
                areaMag.append( np.pi * (5.0**2) )# 0 to Magnitude point radiuses
                
                # sort Events by year ...
                manifestHead = manifestFile.split("."); manifestHead = manifestHead[0]
                if (manifestHead[0:4] == '2005'):
                    evCords2005.append( [eqHeaders['evlo'], eqHeaders['evla'] ] )
                    evMag2005.append(5.0)
                    areaMag2005.append(25)
                elif (manifestHead[0:4] == '2006'):
                    evMag2006.append(5.0)
                    areaMag2006.append(25)
                    evCords2006.append( [eqHeaders['evlo'], eqHeaders['evla'] ] )
                elif (manifestHead[0:4] == '2007'):
                    evCords2007.append( [eqHeaders['evlo'], eqHeaders['evla'] ] )
                    evMag2007.append(5.0)
                    areaMag2007.append(25)
        
            
            # Do polar Plot ......... & Try the Pdf multi-save pages ...
            plotFName = 'StatsOBS/' + DataQual +  "Polar/OBS" +  manifestHead
            plotTitle = "OBS" + manifestHead + " | Total Equakes: " + str( len(manifestTable) ) + " | QC Tag: " + DataQual

            #showPolarEventPlot(evBaz, evDist, areaMag, evMag, plotTitle, plotFName)
            
            
            # Do scatter histogram plot
            plotFName = 'StatsOBS/' + DataQual +  "Histograms/OBS" +  manifestHead
            plotTitle = "OBS" + manifestHead + " | Total Equakes: " + str( len(manifestTable) ) + " | QC Tag: " + DataQual

            #showHistogramScatterPlot(evBaz, evDist, areaMag, evMag, plotTitle, plotFName)
            
            # Draw station Event Map ...
            if (manifestHead[0:4] == '2005'):
                stCords2005.append( [eqHeaders['stlo'], eqHeaders['stla'] ] )
            elif (manifestHead[0:4] == '2006'):
                stCords2006.append( [eqHeaders['stlo'], eqHeaders['stla'] ] )
            elif (manifestHead[0:4] == '2007'):
                stCords2007.append( [eqHeaders['stlo'], eqHeaders['stla'] ] )
    
    
    # Done with all manifests, Plot Maps Now!!
    plotTitle = "2005: " + str( len(stCords2005) ) + " Stations | Earthquakes: " + str( len(evCords2005) / len(stCords2005) ) + " | QC Tag: " + DataQual
    plotFName = 'StatsOBS/' + DataQual +  "Maps/2005Stations"
    showStationEventMap(stCords2005, evCords2005, areaMag2005, evMag2005, plotTitle, plotFName)
    
    #sys.exit()
    
    plotTitle = "2006: " + str( len(stCords2006) ) + " Stations | Earthquakes: " + str( len(evCords2006) / len(stCords2006) ) + " | QC Tag: " + DataQual
    plotFName = 'StatsOBS/' + DataQual +  "Maps/2006Stations"
    showStationEventMap(stCords2006, evCords2006, areaMag2006, evMag2006, plotTitle, plotFName)

    
    plotTitle = "2007: " + str( len(stCords2007) ) + " Stations | Earthquakes: " + str( len(evCords2007) / len(stCords2007) ) + " | QC Tag: " + DataQual
    plotFName = 'StatsOBS/' + DataQual +  "Maps/2007Stations"
    showStationEventMap(stCords2007, evCords2007, areaMag2007, evMag2007, plotTitle, plotFName)
    
    plotTitle =  str( len(stCordsBH) ) + " Stations | Earthquakes: " + str( len(evCordsBH) / len(stCordsBH) ) + " | QC Tag: " + DataQual
    plotFName = 'StatsBorehole/' + DataQual +  "Maps/BoreholeStations"
    showStationEventMap(stCordsBH, evCordsBH, areaMagBH, evMagBH, plotTitle, plotFName)
    
    
    #stCordsAll = stCordsBH + stCords2005
    #evCordsAll = evCordsBH + evCords2005
    #areaMagAll = areaMagBH + areaMag2005
    #evMagAll = 
    #plotTitle =  str( len(stCordsAll) ) + " Stations | Earthquakes: " + str( len(evCordsBH) / len(stCordsBH) ) + " | QC Tag: " + DataQual
    #plotFName = "AllStations"
    #showStationEventMap(stCordsBH, evCordsBH, areaMagBH, evMagBH, plotTitle, plotFName)
    
    

            
        
        
```
