```python
# OLUGBOJI TOLUTLOPE ..... LOAD FINAL RESULTS:
# CALCULATE LAB TEMP @ inferred LAB DEPTH 'H' using Half space cooling model.
# Then generate 3 histograms:
#  1. LAB Depth H
#  2. LAB Temp Thsc 
#  3. LAB % Velocity drop

from matplotlib import rcParams
rcParams['axes.labelsize'] = 14
rcParams['xtick.labelsize'] = 14
rcParams['ytick.labelsize'] = 14
rcParams['legend.fontsize'] = 14

import subprocess
import time, sys, os
import obspyWaveFormCrawler as crwl

from math import *

def TempHSC(zm, tm):
    # return the temperature at a particular depth Zm and for a certain age tm
    
    K = 3.138          # Thermal conductivity (W per m)
    kappa = 0.804e-6  # thermal diffusivity (mSq per sec)
    myr2sec = pi*1e13;
    my2sec = 1e6*3.15e7;
    km2m = 1e3;
    To = 1365

    ModelName = 'HSC [Half Space Cooling] Model';
    TModel = To*erf((zm*km2m) / (2*sqrt(kappa*tm*my2sec)));
    
    return TModel + 273.15

## Test synthetic class ...
manifestDir = "/Users/tmo22/Documents/JapaneseProject/Manifest/"
modelDir = "/Users/tmo22/Documents/JapaneseProject/FwrdModels/"

fileNme = "/Users/tmo22/Documents/JapaneseProject/ManuscriptAndNotes/Results4Distrib/Jap_All.txt"
stNames = np.loadtxt(fileNme, skiprows=1, usecols=[0], dtype='str')
HKresltLAB = np.loadtxt(fileNme, skiprows=1, usecols=[7])

lPulse = np.loadtxt(fileNme, skiprows=1, usecols=[10])
rPulse = np.loadtxt(fileNme, skiprows=1, usecols=[11])
allAge = np.loadtxt(fileNme, skiprows=1, usecols=[3])

print HKresltLAB, stNames

CSmanifest = manifestDir + "CCP_Manifest/CS_Manifest/"
CSmanifestAll = manifestDir + "CCP_Manifest/CS_Manifest_All/"

totGoodEq = []           # List only good earthquakes

manifestFiles = ['T01', 'T02', 'T03', 'T05',  'WP1', 'T06', 'T07', 'T08', 'T09', 'T11', 'T12', 'T13', 'T15', 'T16', 'T18', 'T17', 'T19', 'T21', 'WP2']
manifestFiles = ['T01']

nSta = len(stNames)
nSta =0
manifestFiles = stNames

Depth =  70
Age = 65
#print "Temperature at", Depth, " and ", Age, " is ", TempHSC(Depth, Age)

allTemp = np.loadtxt(fileNme, skiprows=1, usecols=[13])
allVelDrop = np.loadtxt(fileNme, skiprows=1, usecols=[9])
allAni = np.loadtxt(fileNme, skiprows=1, usecols=[12])




#print allTemp
showTemp = []
showVelDrop = []
showAni = []

for iSta in range(nSta, nSta+19):
#for manifestFile in manifestFiles:
    
    #manifestHead = manifestFile.split(".txt")[0]
    manifestHead = manifestFiles[iSta]
    LABDepth = HKresltLAB[iSta]
    Age = allAge[iSta]
    velDrop = allVelDrop[iSta]
    
    #print "Station", manifestHead, "Depth", HKresltLAB[iSta], "Age:", Age, "timing: ", lPulse[iSta], rPulse[iSta]
    
    if LABDepth > 0 and velDrop > 0:
        print "Station", manifestHead, ", Temperature at", LABDepth, " and ", Age, " is ", TempHSC(LABDepth, Age)
        
        showTemp.append( TempHSC(LABDepth, Age) )
        showVelDrop.append( allVelDrop[iSta] )
        showAni.append( allAni[iSta] )
    else:
        print "Station", manifestHead, "Depth not defined"


barTitles = ['Temperature', 'Vel drop' ,'Anisotropy']
print "mean Temp", mean(showTemp)
print "mean Vel Drop", mean(showVelDrop)



fig = plt.figure(figsize = (3,9) )


# Plot Temperature
ax = plt.subplot2grid((3,1), (0,0))
y, binEdges = np.histogram(showTemp)
bincenters = 0.5*(binEdges[1:]+binEdges[:-1])
plt.plot(bincenters, y, '-',linewidth=2, color='k')
plt.fill_between(bincenters, 0, y, y >=  0, color='r', alpha=0.4)

plt.scatter(bincenters, y)
plt.ylim([0, 4])
plt.xlim([920, 1401])
ax.set_xlabel(r"$T (^\circ K)$")
ax.set_ylabel("No. of Stations")

# Plot Vel Drop
ax = plt.subplot2grid((3,1), (1,0))
y, binEdges = np.histogram(showVelDrop)
bincenters = 0.5*(binEdges[1:]+binEdges[:-1])
plt.bar(bincenters, y, alpha=0.4, color = 'r')
ax.set_xlabel(r"$\frac{\Delta Vs}{Vs} (\%)$")
ax.set_ylabel("No. of Stations")

# Plot Anisotropy
ax = plt.subplot2grid((3,1), (2,0))
y, binEdges = np.histogram(showAni)
bincenters =  0.5*(binEdges[1:]+binEdges[:-1]) 
plt.bar(bincenters, y, alpha=0.4, color='r')
ax.set_xlabel("Anisotropy")
ax.set_ylabel("No. of Stations")

fig.subplots_adjust(hspace=0.3)

format = 'ps'
FINALRESULTSFILE = "/Users/tmo22/Documents/JapaneseProject/Images/ManuscriptResults/Histograms"
plt.savefig(FINALRESULTSFILE+'.'+format, format=format)

plt.show()
```
